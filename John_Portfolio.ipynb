Database Implementation and Insight Generation - Recycle4Change

This project is dedicated to producing a robust database capable of accurately storing information about Recycle4Change, a program in Australia focused on recycling and sustainability. My goal is to develop comprehensive queries and reports that provide valuable data insights to drive informed decisions.

Project Overview

This project includes the following components:

1. ER Model
A detailed Entity-Relationship (ER) model representing the logical structure of the Recycle4Change database.

2. SQL Commands
- Scripts to physically implement the logical model: Transforming ER models into a fully functioning database schema (see SQL database output below).
- Scripts to populate the database: Inserting sample data to set up the database for operational use.
- Scripts to retrieve meaningful information: Queries designed to extract insightful data, aiding in business analysis and decision-making.

3. Visualisations
Data visualisations that provide actionable insights from the database, helping to understand trends and patterns.

Tools and Techniques Used: 
- MySQL Database Engine
- Power BI Visualisation
- Data Modelling
- Data Mining
- Insight Generation
